```python
%timeit (df1 < df2) & (df2 <= df3) & (df3 != df4)
```

```python
def create_df(cols, idx):
    data = {c : [str(c.lower()) + str(i) for i in idx] for c in cols}
    return pd.DataFrame(data, idx)
```

##### 리샘플링 (Resampling)
- 리샘플링 : 시계열의 빈도 변환
- 다운샘플링 : 상위 빈도 데이터를 하위 빈도 데이터로 집계
- 업샘플링 : 하위 빈도 데이터를 상위 빈도 데이터로 집계
- resample method




```python
dr = pd.date_range('20200101', periods=200, freq='D')
ts = pd.Series(np.random.randn(len(dr)), index = dr)
ts
```

```python
ts.resample('M').mean() # 월별 평균을 마지막날을 index로 하여 return
```

```python
ts.resample('M', kind = 'period').mean() # 주기에 따라 resampling
```

```python
dr = pd.date_range('20200101', periods=10, freq='T')
ts = pd.Series(np.arange(len(dr)), index = dr)
ts
```

```python
ts.resample('2T', closed = 'left').sum()
```

```python
ts.resample('2T', closed = 'right').sum()
```

```python
ts.resample('2T', closed = 'right', label='right').sum() # label로 기준점을 다시 되돌림
```

```python
ts.resample('2T', closed = 'right', label='right', offset='-1s').sum() # -1초씩 조정
```

```python
ts.resample('2T').ohlc()
```

무빙 윈도우(Moving Window)

```python
df = pd.DataFrame(np.random.randn(300, 4),
                  index = pd.date_range('20200101', periods=300, freq='D'),
                  columns=['C1', 'C2', 'C3', 'C4'])
df
```

```python
df.rolling(30).mean().plot() # 30일간격으로 이동하는 plot을 그려줌
```

```python
df.rolling(60).mean().plot() # 60일간격으로 이동하는 plot을 그려줌
```

```python
df.C1.rolling(60, min_periods=10).std().plot()
```

```python
df.rolling(60, min_periods=10).std()[10:50].plot()
```

```python
df.rolling(60, min_periods=10).std().expanding().mean().plot() # expanding : 보는 구간을 점점 확장시켜줌
```

```python
df.rolling(60).mean().plot(logy = True) # log scale로 바꿔줌
```

```python
df.rolling('20D').mean().plot()
```

```python
df.C1.rolling(30, min_periods=20).mean().plot(style = '--', label = 'Simple MA')
df.C1.ewm(span = 30).mean().plot(style = '-', label = 'EWMA') # EWMA : rolling과 expanding을 같이 사용하는 방법
```

```python
df.C1.rolling(100, min_periods=50).corr(df.C3).plot() # C1과 C3의 correlation
```
